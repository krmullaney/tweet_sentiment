```python
train = pd.read_csv('train_tweets.csv')
test = pd.read_csv('test_tweets.csv')
train_reply1 = pd.read_csv('train_reply1_tweets.csv')
train_reply2 = pd.read_csv('train_reply1_tweets.csv')
```

```python
# Source: https://medium.com/analytics-vidhya/pre-processing-tweets-for-sentiment-analysis-a74deda9993e

def preprocess(input_df):
    # Make copy of input data frame
    df = input_df.copy()

    # Get rid of punctuation
    df['text'] = df['text'].str.replace('[{}]'.format(string.punctuation), '')

    # Convert to lowercase
    df['text'] = df['text'].str.lower()

    # Get rid of URL links
    df['text'] = df['text'].apply(lambda x: re.sub(r'https?:\/\/\S+', '', x))
    df['text'] = df['text'].apply(lambda x: re.sub(r"www\.[a-z]?\.?(com)+|[a-z]+\.(com)", '', x))

    # Remove videos and links
    df['text'] = df['text'].apply(lambda x: re.sub(r'{link}', '', x))
    df['text'] = df['text'].apply(lambda x: re.sub(r"\[video\]", '', x))

    # Handle HTML reference characters
    df['text'] = df['text'].apply(lambda x: re.sub(r'&[a-z]+;', '', x))

    # Remove non-letter characters
    df['text'] = df['text'].apply(lambda x: re.sub(r"[^a-z\s\(\-:\)\\\/\];='#]", '', x))

    # Get rid of twitter handles
    df['text'] = df['text'].apply(lambda x: re.sub(r'@mention', '', x))

    return df
```

```python
train_clean = preprocess(train)
train_clean['text'][:10]
```

```python
test_clean = preprocess(test)
test_clean['text'][:10]
```

```python
train_reply1_clean = preprocess(train_reply1)
train_reply1_clean['text'][:10]
```

```python
train_reply2_clean = preprocess(train_reply2)
train_reply2_clean['text'][:10]
```
